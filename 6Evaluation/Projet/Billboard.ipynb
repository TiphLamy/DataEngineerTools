```
import requests
from bs4 import BeautifulSoup
```

```
import pandas as pd
```

```
url = "https://www.billboard.com/charts/billboard-200"
```

```
response = requests.get(url)
soup = BeautifulSoup(response.text)
```

```
#Semaine du 07/12/19
Artistes = []
Titres = []
Last_Week = []
Peak = []
Duration = []
Progression = []


artistes = soup.find_all(class_="chart-element__information__artist text--truncate color--secondary")[0:200]
titres = soup.find_all(class_="chart-element__information__song text--truncate color--primary")[0:200]
last_Week = soup.find_all(class_="chart-element__meta text--center color--secondary text--last")[0:200]
peak = soup.find_all(class_="chart-element__meta text--center color--secondary text--peak")[0:200]
duration = soup.find_all(class_="chart-element__meta text--center color--secondary text--week")[0:200]
progression = soup.find_all(class_="chart-element__information__delta__text text--default")[0:200]

for i in range(200):
  Artistes.append(artistes[i].text)
  Titres.append(titres[i].text)
  Last_Week.append(last_Week[i].text)
  Peak.append(peak[i].text)
  Duration.append(duration[i].text)
  Progression.append(progression[i].text)

Billboard_200 = pd.DataFrame({ 'Artistes' : Artistes,
                               'Titres' : Titres,
                               'Last Week' : Last_Week,
                               'Peak' : Peak,
                               'Duration' : Duration,
                               'Progression' : Progression
                               })
 #Billboard_200.index.name = 'Rank'
Billboard_200.index +=1
```
